```python
#statistics of sepal length
print("Mean of sepal length: "+ str(np.mean(dataset['SepalLengthCm'])))

print("Median of sepal length: ", np.median(dataset['SepalLengthCm']))

print("Mode of sepal length:: ", dataset['SepalLengthCm'].mode())

print ("Standard deviation of sepal length: ", np.std(dataset['SepalLengthCm']))
```

```python
plt.hist(dataset["SepalWidthCm"], color= "g")
plt.title("A histogram showing distribution of Sepal Width in cm")
plt.xlabel("Sepal Width(cm)")
plt.ylabel("Count")
plt.show()
```

```python
plt.hist(dataset["SepalLengthCm"], color= "b")
plt.title("A histogram showing distribution of Sepal Length in cm")
plt.xlabel("Sepal Length(cm)")
plt.ylabel("Count")
plt.show()
```

```python
plt.scatter(x= dataset["PetalWidthCm"], y= dataset["PetalLengthCm"], color= "g")
plt.title("A scatter plot showing correlation between petsl width and petal length in cm")
plt.xlabel("Petal Width(cm)")
plt.ylabel("Petal Length(cm)")
plt.show()

print(np.corrcoef(dataset['PetalWidthCm'], dataset["PetalLengthCm"]))
```

therefore, there is a positive correlation between peeal length and petal width. that means a longer petal is also having a higher value of length. 

```python
plt.scatter(x= dataset["SepalWidthCm"], y= dataset["SepalLengthCm"], color= "r")
plt.title("A histogram showing a comparison between Sepal Length and Sepal Width")
plt.xlabel("Sepal Width(cm)")
plt.ylabel("Sepal Length(cm)")
plt.show()
print(np.corrcoef(dataset['SepalWidthCm'], dataset["SepalLengthCm"]))
```

There appears to be a slight negative correlation between sepal length and width with  an increase in sepal width overall being related to a decreasing sepal length.
